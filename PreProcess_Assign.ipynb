## Outlier Removal

 outlier detection analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cols_to_plot = ['Living Area after Renovation (in Sqft)', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)', 'Age of House (in Years)','Basement Area (in Sqft)','Lot Area after Renovation (in Sqft)']
```

Except for Age of house(in years) there are large no of outliers in major cols

```python
cols_to_fix = ['Living Area after Renovation (in Sqft)', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)','Basement Area (in Sqft)']
```

```python
def detect_fix_outliers(df, columns):
    outlier_indices = []

    for col in columns:

        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # Bounds
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR

        #Outliers
        outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
        # Collect indices of rows with outliers
        outlier_indices.extend(outliers.index.tolist())

        print(f"--- Column: {col} ---")
        print(f"Lower Bound: {lower_limit:.2f} | Upper Bound: {upper_limit:.2f}")
        print(f"Total Outliers Detected: {len(outliers)}\n")
        print(f"Combined Outliers: {list(set(outlier_indices))} \n")
        print(f"total no of Combined Outliers: {len(list(set(outlier_indices)))} \n")
        #Capping
        df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)

  #  return list(set(outlier_indices))
    return df
```

```python
df_capped = detect_fix_outliers(hp_data, cols_to_fix)
```
